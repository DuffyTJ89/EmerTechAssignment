The Iris dataset is made up of three different types of irises, namely Setosa, Versicolour and Virginica, with their petal and sepal length stored in an array. This notebook will aim to explain the dataset and cover the difficulty in writing an algorithm to seperate the three classes of iris mentioned above based on the variables in the dataset.

```python
# Import pandas. Provides conversion to and from pamdas data structures
import pandas as pd

# Load the iris data set from a URL.
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")

# Display the dataset
df
```

```python
# return the Id and the type of iris it is
df['species']
```

```python
df[['petal_length', 'species']]
```

Already we can see that there are differences in petal length between the species of iris but as we'll see with further investagation this alone isn't enough to distinguish them. 

```python
# turn off a warning message
import warnings
warnings.filterwarnings('ignore')

# Seaborn is a python data visualization library based on matplotlib.
import seaborn as sns
import matplotlib.pyplot as plt


# Display graphs to show the overlap between the different species

sns.pairplot(df, hue='species')
plt.show()
```

```python
# these graphs show just how similar the width and length can be with versicolor and virginica
sns.violinplot(y='species', x='sepal_length', data=df, inner='quartile')
plt.show()
sns.violinplot(y='species', x='sepal_width', data=df, inner='quartile')
plt.show()
sns.violinplot(y='species', x='petal_length', data=df, inner='quartile')
plt.show()
sns.violinplot(y='species', x='petal_width', data=df, inner='quartile')
plt.show()
```

ref: {https://www.kaggle.com/jchen2186/machine-learning-with-iris-dataset }

So as we see the setosa are in most cases fairly different to the other two species. The problem is seperating the versicolor and virginica so the can be easily identifiable. There is a lot of overlap in some cases and this makes it difficult to quickly identify them from the graphs.

For this reason the iris dataset is really useful for testing machine learning as there are two species that are very similar in a lot of respects so when shown to a human in a graph it isn't immediately obvious which is which. But we can use machine learning to solve this problem by teaching a neural network to understand the differences between the species.
